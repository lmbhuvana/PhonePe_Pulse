```python
# To import packages and libraries
import pandas as pd
import json
import os
import mysql.connector as sql
import streamlit as st
```

```python
# To get statewise list from aggregated_transaction 

path=r"C:\Users\Bhuvana Vignesh\OneDrive\Desktop\GUVI_Projects\PhonePe_Pulse\pulse-master\pulse-master\data\aggregated\transaction\country\india\state"
agg_trans_state_list=os.listdir(path)

data={'State':[], 'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

for i in agg_trans_state_list:
    p_i=path+"/"+i
    agg_year=os.listdir(p_i)

    for j in agg_year:
        p_j=p_i+"/"+j
        agg_year_list=os.listdir(p_j)

        for k in agg_year_list:
            p_k=p_j+"/"+k
            Data=open(p_k,'r')
            D=json.load(Data)

            for z in D['data']['transactionData']:
                Name=z['name']
                count=z['paymentInstruments'][0]['count']
                amount=z['paymentInstruments'][0]['amount']
                data['State'].append(i)
                data['Year'].append(j)
                data['Quarter'].append('Q'+str(k[0]))
                data['Transaction_type'].append(Name)
                data['Transaction_count'].append(count)
                data['Transaction_amount'].append(amount)
                

df_agg_trans_state_list=pd.DataFrame(data)
df_agg_trans_state_list.to_csv('aggregated_transaction.csv',index=False)
# df_agg_trans_state_list.isnull().sum()
```

```python
# To get statewise User details from aggregated_users
import os
path=r"C:\Users\Bhuvana Vignesh\OneDrive\Desktop\GUVI_Projects\PhonePe_Pulse\pulse-master\pulse-master\data\aggregated\user\country\india\state"
agg_user_state_list=os.listdir(path)

data={'State':[],'Year':[], 'Quarter':[],'Brands':[],'Count':[],'Percentage':[]}

for i in agg_user_state_list:
    p_i=path+"/"+i
    agg_year=os.listdir(p_i)

    for j in agg_year:
        p_j=p_i+"/"+j
        agg_year_list=os.listdir(p_j)

        for k in agg_year_list:
            p_k=p_j+"/"+k
            Data=open(p_k,'r')
            D=json.load(Data)

            if D['data']['usersByDevice'] is not None:
                for z in D['data']['usersByDevice']:
                    Brand1=z['brand']
                    Count1=z['count']
                    Percentage1=z['percentage']
                    data['Brands'].append(Brand1)
                    data['Count'].append(Count1)
                    data['Percentage'].append(Percentage1)
                    data['State'].append(i)
                    data['Year'].append(j)
                    data['Quarter'].append('Q'+str(k[0]))
df_agg_user_state_list=pd.DataFrame(data)
df_agg_user_state_list.to_csv('aggregated_user.csv',index=False)
# df_agg_user_state_list.isnull().sum()              
```

```python
# To get statewise transaction from Map
import os
path = r"C:\Users\Bhuvana Vignesh\OneDrive\Desktop\GUVI_Projects\PhonePe_Pulse\pulse-master\pulse-master\data\map\transaction\hover\country\india\state"
map_state_list = os.listdir(path)

data = {'State': [], 'Year': [], 'Quarter': [], 'District': [],'Transaction_count': [], 'Transaction_amount': []}

for i in map_state_list:
    path_i = path + "/"+i
    map_year_list = os.listdir(path_i)

    for j in map_year_list:
        path_j = path_i +"/"+ j
        map_year_json = os.listdir(path_j)

        for k in map_year_json:
            path_k = path_j +"/"+ k
            Data = open(path_k, 'r')
            D = json.load(Data)

            for z in D['data']['hoverDataList']:
                district = z['name'].split('district')[0]
                count = z['metric'][0]['count']
                amount = z['metric'][0]['amount']

                data['State'].append(i)
                data['Year'].append(j)
                data['Quarter'].append('Q'+str(k[0]))
                data['District'].append(district)
                data['Transaction_count'].append(count)
                data['Transaction_amount'].append(amount)
df_map_state_list=pd.DataFrame(data)
df_map_state_list.to_csv('map_transaction.csv', index=False)
# df_map_state_list.isnull().sum()                
```

```python
# To get statewise User details from Map

path=r"C:\Users\Bhuvana Vignesh\OneDrive\Desktop\GUVI_Projects\PhonePe_Pulse\pulse-master\pulse-master\data\map\user\hover\country\india\state"
map_user_state_list=os.listdir(path)

data = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Registered_user': [], 'App_opens': []}

for i in map_user_state_list:
    path_i = path +"/"+ i
    map_user_year = os.listdir(path_i)

    for j in map_user_year:
        path_j = path_i +"/"+ j
        map_user_year_list = os.listdir(path_j)

        for k in map_user_year_list:
            path_k = path_j +"/"+ k
            Data= open(path_k, 'r')
            D= json.load(Data)

            for z_key, z_value in D['data']['hoverData'].items():
                district = z_key.split(' district')[0]
                reg_user = z_value['registeredUsers']
                app_opens = z_value['appOpens']

                data['State'].append(i)
                data['Year'].append(j)
                data['Quarter'].append('Q'+str(k[0]))
                data['District'].append(district)
                data['Registered_user'].append(reg_user)
                data['App_opens'].append(app_opens)
df_map_user_state_list=pd.DataFrame(data)
df_map_user_state_list.to_csv('map_user.csv', index=False)
# df_map_user_state_list.isnull().sum()
```

```python
#To get districtwise transaction from Top

path=r"C:\Users\Bhuvana Vignesh\OneDrive\Desktop\GUVI_Projects\PhonePe_Pulse\pulse-master\pulse-master\data\top\transaction\country\india\state"
top_trans_state_list=os.listdir(path)

data = {'State': [], 'Year': [], 'Quarter': [], 'District_Pincode': [], 'Transaction_count': [], 'Transaction_amount': []}

for i in top_trans_state_list:
            path_i = path +"/"+ i
            top_year_list = os.listdir(path_i)

            for j in top_year_list:
                path_j = path_i + "/"+j
                top_year_json = os.listdir(path_j)

                for k in top_year_json:
                    path_k = path_j +"/"+ k
                    f = open(path_k, 'r')
                    d = json.load(f)

                    for z in d['data']['districts']:
                        district_code = z['entityName']
                        count = z['metric']['count']
                        amount = z['metric']['amount']

                        data['State'].append(i)
                        data['Year'].append(j)
                        data['Quarter'].append('Q'+str(k[0]))
                        data['District_Pincode'].append(district_code)
                        data['Transaction_count'].append(count)
                        data['Transaction_amount'].append(amount)
df_top_trans_state_list=pd.DataFrame(data)
df_top_trans_state_list.to_csv('top_transaction.csv', index=False)
# df_top_trans_state_list.isnull().sum()              
```

```python
# To get postalcodewise Top lists

path=r"C:\Users\Bhuvana Vignesh\OneDrive\Desktop\GUVI_Projects\PhonePe_Pulse\pulse-master\pulse-master\data\top\user\country\india\state"
top_trans_state_postal=os.listdir(path)

data = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Registered_Users':[]}

for i in top_trans_state_postal:
    path_i = path + '/'+i
    top_year_list = os.listdir(path_i)

    for j in top_year_list:
        path_j = path_i + '/'+j
        top_year_list = os.listdir(path_j)

        for k in top_year_list:
            path_k = path_j + '/'+k
            Data = open(path_k, 'r')
            D = json.load(Data)

            for z in D['data']['pincodes']:
                pincode = z['name']
                registeredUser=z['registeredUsers']

                data['State'].append(i)
                data['Year'].append(j)
                data['Quarter'].append('Q'+str(k[0]))
                data['Pincode'].append(pincode)
                data['Registered_Users'].append(registeredUser)
df_top_trans_state_postal=pd.DataFrame(data)
df_top_trans_state_postal.to_csv('top_postalcode.csv', index=False)
df_top_trans_state_postal.isnull().sum()

# missing_pincode_rows = df_top_trans_state_postal[df_top_trans_state_postal['Pincode'].isnull()]

# Display the rows with missing 'pincode'
# print(missing_pincode_rows)

# mean_pincode = df_top_trans_state_postal['Pincode'].mean()
# mean_pincode
# df_top_trans_state_postal['Pincode'].fillna(mean_pincode, inplace=True)
```
